Two inputs are required for the KMeans method: a DataFrame of numeric attributes and a number of clusters to generate. Because we have an ID column, PatientID, which is both non-numeric and not useful in identifying cluster groupings, we will use column indexing ([1:9]) to exclude the first column of the HeartRisk DataFrame from our analysis. We will add some other, optional arguments to the KMeans() method as well: max_iter limits the number of times R will loop through the data, re-classifying observations into the different clusters seeking for the greatest possible degree of similarity within and between groups. We will set this to 10 iterations. The argument random_state determines how observations are initially selected into groups for classification. We do not want a random state because this makes replication of results difficult, so we will set this to 0. And finally, n_init allows us to set a specific seed number to start the process of classification, making replication of results easier. This can be set to any number, but for consistency we will use 7. Use the “+ Code” button to add a new command prompt to your Notebook. Enter the command below and then execute the code:

```python
from sklearn.cluster import KMeans
HeartKMeans = KMeans(n_clusters=3, n_init=7, max_iter = 10, random_state=0).fit(HeartRisk.iloc[:, np.r_[1:9]])
```

Now we will use the HeartKMeans model to create a centroid table. This table will hold the size of each cluster, which is the number of observations grouped into each one. It will also list the centers, or means, for each attribute across all clusters. This table is particularly useful when determining what labels you will assign to each cluster. This will be discussed further in the Evaluation section. Two “for” loops in Python are employed here: one to display the sizes of the clusters and one to display the means for the attributes for each cluster. Add the following code to the Console and press Enter. Python numbers clusters starting with 0. The #comments to the right of the code provide additional explanation of each command’s function.

```python
for i in range(len(HeartKMeans.cluster_centers_)):
  print("Cluster", i, "-- Size:", sum(HeartKMeans.labels_ == i))    #Sum up the number of observations in each cluster and print the results
HeartCentroid = [HeartKMeans.feature_names_in_]     #Create an array with the attribute names defining the number of columns in the array
for p in range(len(HeartKMeans.cluster_centers_)):
    HeartCentroid.append(HeartKMeans.cluster_centers_[p])  #Put the means for each attribute into 'k' number of rows in the array, one for each cluster
HeartCentroid = pd.DataFrame(HeartCentroid)    #Change the array into a pandas DataFrame for easier readability
HeartCentroid.columns = HeartCentroid.iloc[0]
HeartCentroid = HeartCentroid[1:]  #Change the DataFrame's first row into column headers composed of the attribute names
HeartCentroid = HeartCentroid.reset_index()     #Re-number the clusters starting from zero after changing the first row to column headers
HeartCentroid = HeartCentroid.drop('index', axis=1)    #Drop the old row numbering
HeartCentroid
```

Finally, we will want to know which insurance customers (PatientIDs) were placed into each cluster. Add a new command prompt to your Notebook using “+ Code” and execute the command below. This will create a new copy of the HeartRisk DataFrame called RiskClusters that will include each observation’s cluster assignment in a new column called Cluster_Number. The rightmost column of this new DataFrame will display the cluster for each observation.

```python
RiskClusters = HeartRisk      #Creates a copy of the HeartRisk data in a new DataFrame called RiskClusters
RiskClusters['Cluster_Number']=HeartKMeans.labels_        #Adds the cluster number of each observation in RiskCluster
pd.set_option('display.max_rows', None)          #Ensures each observation will be shown in the results
RiskClusters          #Prints out the DataFrame containing each observation and its cluster assignment
```

People sometimes wonder: How many clusters do I need? What is the right value for k? There is no one right answer to these questions; however, there are some techniques that may be helpful in determining a reasonable number of clusters for a given data set. It is important to remember that the ideal number of clusters for each data set will be different depending on the number of attributes, the variability in the values within and among the attributes, and the size of the data set. There are several techniques to help analysts identify an optimal number of clusters for a given data set. These include visualizations such as scatterplots to look for groupings that may appear across correlated attributes, along with algorithmic approaches such as the Elbow Method, Silhouette Method, and Gap Statistics. Ultimately, the right number of clusters for an analysis is the number that yields the greatest insights into the data. We will discuss this more in the next section: Evaluation.

Sonia’s major objective in the hypothetical scenario posed in this chapter was to try to find groupings of heart disease risk among insurance customers. Using the k-Means technique, we have identified three clusters for Sonia, and we can now evaluate their usefulness in addressing her question. We will start by looking at our centroid table. Each data analytics software product will number clusters differently, but no software will apply labels, such as “Low Risk,” to the clusters. This is the job of the data analyst. The meaning of the values in the clusters is up to us to determine.

Because various analytics software products will display k-means clustering results differently, we will examine the clusters based on the values they contain, rather than by assigned cluster number. Examine Figure 6.16. This is a centroid table as displayed in the RStudio software. Look at the cluster on the top row of the table. There are 180 individuals in this cluster. We might label this cluster as “Low Risk Customers.” Why? Primarily because this cluster has the lowest averages for major indicators of heart disease risk. This group has the lowest average weight, cholesterol, and age; lowest incidence of diabetes, previous heart events, and exercise-induced issues; and lowest resting heart rate. Thus, these health insurance customers are probably at the lowest risk of developing heart problems, at least as they relate to the attributes in this data set. Notice that the average in the Sex variable for this cluster is not the lowest in the table. In this attribute, a value of 0 represents Female and 1 represents Male, so a mean of 0.55 indicates that there are more men than women represented in this cluster.

An examination of the Sex variable across all three clusters shows that only one of the three groups has more females than males in it. The cluster containing 206 people has an average Sex of 0.41. This indicates that there are more zeros (0 = female) than ones (1 = male) in this cluster. Filtering of the data set on assigned cluster number and sex reveals that there are 121 females and 85 males in this cluster. We can determine that 121 zeros plus 85 ones divided by 206 people equals an average Sex value of 0.41. Put another way, approximately 41% of people in this cluster are male and 59% are female. This cluster might be labeled as “Moderate Risk Customers,” since the means across all of the attributes are between the values in the other two clusters. To cite two examples: Cholesterol = 217.1, which falls between 158.0 and 259.3 in the other two clusters, and Age = 55.9, which falls between 45.9 and 66.6 in the other two clusters.

Finally, let’s examine the smallest cluster, the one with 161 individuals in it. The risk factors have the highest averages across all attributes in this cluster. Look at the means for the Diabetic attribute. In this cluster, which we might label “High Risk Customers,” the Diabetic value is almost double the values in both of the other clusters. What might we derive from this observation? First, in the low and moderate risk clusters, there are more people who do not have a diabetic condition than those who do. But in this highest risk group, nearly 80% of people have a diabetic condition. We can surmise from this that diabetes is correlated with other heart disease risk factors, and thus, any interventions that may be developed should probably include diagnosis and treatment for this particular issue.

These are just some insights that can be derived from our initial k-Means analysis of this data. You may spot other interesting comparison values.
